# Unicycle Model - Infeasible Start
One of the disadvantages of using a shooting-based method like iLQR is that we can only provide a dynamically feasible initial guess. In practice, this means we can really only provide an initial guess for the controls sequence. However, for many problems, providing an initial infeasible guess for the state trajectory is highly desireable, for instance when using a sample-based planner to warm-start your trajectory optimization solver.

With a constrained solver like ALTRO, we can allow the user to provide an initial dynamically-infeasible guess through simple state augmentation. If our original nonlinear, underactuated discrete dynamics take the form:
$$ x_{k+1} = f(x_k,u_k) $$

with $x \in \mathbb{R}^b$ and $u \in \mathbb{R}^m$, we can add $n$ controls $\hat{u} \in \mathbb{R}^n$ such that the new dynamics become
$$ x_{k+1} = f(x_k, u_k) + \hat{u} $$
which makes the new system trivially fully-actuated. When given a dynamically infeasible initial trajectory for the original system $\bar{x}_{1:N}, \bar{u}_{1:N-1}$, we simply calculate the additional controls $\hat{u}$ as 
$$ \hat{u} = \bar{x}_{k+1} - f(\bar{x}_k, \bar{u}_k) $$
and then add the constraint that $\hat{u} = 0$ to the solver. Using augmented Lagrangian, as long as the initial penalty on this constraint isn't too high, it will gradually decrease these additional controls until they are zero and we recover a dynamically-feasible solution by the end of the solve.

## 1. Problem Set Up
To demonstrate this approach using ALTRO, which can do this for us automatically, we'll look at a fairly simple obstacle avoidance problem using the standard bicycle model. To keep things simple, we'll use the model from RobotZoo.jl

### 1.1 Model and Objective

```julia
# Model
model = RobotZoo.DubinsCar()
n,m = size(model)
x0 = @SVector [2.5,2.5,0.]
xf = @SVector [7.5,2.5,0.]
N = 101    # number of knot points
tf = 3.0   # final time (s)

# Cost
Q = (1e-3)*Diagonal(@SVector ones(n))
R = (1e-2)*Diagonal(@SVector ones(m))
Qf = 100.0*Diagonal(@SVector ones(n))
obj = LQRObjective(Q,R,Qf,xf,N);
```

### 1.2 Constraints

```julia
# Obstacle constraints
r = 0.5
s1 = 30; s2 = 50; s3 = 15

circles_escape = NTuple{3,Float64}[]

for i in range(0,stop=5,length=s1)
    push!(circles_escape,(0.,i,r))
end
for i in range(0,stop=5,length=s1)
    push!(circles_escape,(5.,i,r))
end
for i in range(0,stop=5,length=s1)
    push!(circles_escape,(10.,i,r))
end
for i in range(0,stop=10,length=s2)
    push!(circles_escape,(i,0.,r))
end
for i in range(0,stop=3,length=s3)
    push!(circles_escape,(i,5.,r))
end
for i in range(5,stop=8,length=s3)
    push!(circles_escape,(i,5.,r))
end

n_circles_escape = 3*s1 + s2 + 2*s3

circles_escape
x,y,r = collect(zip(circles_escape...))
x = SVector{n_circles_escape}(x)
y = SVector{n_circles_escape}(y)
r = SVector{n_circles_escape}(r)

obs = CircleConstraint(n,x,y,r)
n_circles_escape
```

```julia
# Other constraints
bnd = BoundConstraint(n,m,u_min=-5.,u_max=5.)
goal = GoalConstraint(xf);
```

### 1.3 Build the Problem

```julia
conSet = ConstraintList(n,m,N)
add_constraint!(conSet, obs, 2:N-1)
add_constraint!(conSet, bnd, 1:N-1)
add_constraint!(conSet, goal, N:N)

U0 = [@SVector ones(m) for k = 1:N-1]

prob = Problem(model, obj, xf, tf; constraints=conSet, x0=x0)
initial_controls!(prob, U0);
```

### 1.4 Visualize the Problem
It's always helpful to have a visualization! Let's use some convenient methods in the TrajOptPlots package to visualize the circle constraints and our goal constraints. Note that this really only works well for 2D problems. For 3D problems, we'll use the `visualize!` methods that work with the 3D renderer MeshCat.

```julia
# load plotting packages
using Plots
using TrajOptPlots
pyplot()   # use the pyplot / matplotlib backend
```

```julia
plot(prob)  # yes, it's that easy!
```

### 1.5 Design an initial guess
Now let's design an initial guess to get us from the room on the left to our goal on the right. You can do this however you want, but we'll just read some points off of the plot above. We picked 6 points:

```julia
X_guess = [2.5 2.5 0.;
           4. 5. .785;
           5. 6.25 0.;
           7.5 6.25 -.261;
           9 5. -1.57;
           7.5 2.5 0.]
scatter!(X_guess[1:end-1,1], X_guess[1:end-1,2], color=:gray, markerstrokecolor=:gray, label="")
```

But we need to provide a guess for the entire trajectory, so let's interpolate between these points using a 3rd-order spline. Altro provides a convenient method for doing this, based on the [Interpolations.jl](https://github.com/JuliaMath/Interpolations.jl) package.

```julia
# Use a 3rd order spline to interpolate the rows independently
X0_escape = Altro.interp_rows(N,tf,Array(X_guess'))

# Provide the guess to the problem
initial_states!(prob, X0_escape)

# Plot it
traj2!(states(prob), color=:gray)
# OR
plot(prob)
```

## 2. Solve the Problem
Now that we've designed our problem and our initial, dynamically infeasible initial trajectory, let's solve it using the technique we described in the introduction. As mentioned, Altro.jl can do this automatically if we ask it to. All we have to do is pass `infeasible = true` to the constructor for the `ALTROSolver`, an (optionally) a value for `R_inf` which is the penalty cost for the additional controls. The solution can be a fairly sensitive to this parameter, so it's worth playing around with.

```julia
opts = SolverOptions(
    constraint_tolerance=1e-5,
    cost_tolerance_intermediate=1e-3,
    penalty_scaling=100.,
    penalty_initial=10.,
    active_set_tolerance_pn=1e-4,
    active_set_tolerance_al=1e-4,
)
initial_controls!(prob, U0)
initial_states!(prob, X0_escape)
solver = ALTROSolver(prob, opts, infeasible=true, R_inf=0.1)
solve!(solver);
```

## 3. Plot the solution
Let's plot the solution to see the trajectory the solver found. You will notice that it gets arbitrarily close to the walls, so make sure to add a reasonably-sized buffer to your obstacles when treating obstacles as hard constraints.

```julia
plot(prob)
traj2!(states(solver), label="solution", color=:blue)
```

## 4. Projecting back onto Original Dynamics
Since we've augmented our dynamics with artificial controls, our solution will be slightly dynamically infeasible, depending on how small those extra controls are. However, we can use iLQR to "project" our solution onto the nearest dynamically feasible one. We do this by creating an iLQR solver and running a single iteration of a backwards and forwards pass. Since the slack controls for our problem are very small, it actually doesn't make any difference, but this following trick can be useful when applying this to other systems.

```julia
# Divide the controls
Usol = hcat(Vector.(controls(solver))...)[1:size(prob)[2], :]        # extract out the actual controls
Uextra = hcat(Vector.(controls(solver))...)[size(prob)[2]+1:end, :]  # extract out the additional controls if you want to look at them (ours are very small)

# Copy the solution back to the original problem, discarding the extra controls
initial_states!(prob, states(solver))
initial_controls!(prob, Usol)

# Create a new iLQR solver
ilqr = Altro.iLQRSolver(prob)

# Find a locally-optimal feedback policy
Altro.backwardpass!(ilqr)

# Make the system dynamically feasible by simulating the system forward with feedback (no feed-forward) control
rollout!(ilqr, 0.0)
```
